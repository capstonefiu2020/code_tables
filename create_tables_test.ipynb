### CREATE BIDLO TABLE

```python
df_bidlo = df[['date', 'symbol', 'bidlo']]
df_bidlo.head(1)
```

```python
df_temp = df_bidlo[df_bidlo['symbol']=='A']
df_temp.index=df_temp['date']
df_temp = df_temp[['bidlo']]
df_temp = df_temp.T
df_temp.index = ['A']
df_temp.index = df_temp.index.rename('symbol')
df_temp.head()
```

```python
# df_temp.index
# df_temp.columns
```

```python
df_bidlo_temp = pd.DataFrame([])
for symbol in list_symbol:
    df_temp = df_bidlo[df_bidlo['symbol']==symbol]
    df_temp.index=df_temp['date']
    df_temp = df_temp[['bidlo']]
    df_temp = df_temp.T
    df_temp.index = [symbol]
    #df_temp.rename()
    df_bidlo_temp = pd.concat([df_bidlo_temp, df_temp], axis=0)

df_bidlo_temp
```
